```python
import pandas as pd
import re
import requests as rq
from bs4 import BeautifulSoup
import json
```

# Name of Cluster

```python
req = rq.get("https://secure1.rochester.edu/registrar/api/clusters/departments")
html = req.text
soup = BeautifulSoup(html, "lxml")
```

```python
cluster = json.loads(soup.find("p").text)
clusters = cluster["results"]
```

```python
HUM = []
NSE = []
SSC = []
for num in range(0,len(clusters)):
    if clusters[num]["division"] == "HUM":
        HUM.append(clusters[num]["code"])
    elif clusters[num]["division"] == "NSE":
        NSE.append(clusters[num]["code"])
    else:
        SSC.append(clusters[num]["code"])
    
```

```python
Clusters_division ={"HUM":HUM, "NSE":NSE, "SSC":SSC}
```

# Cluster 

```python
def Cluster(division, code):
    url = "https://secure1.rochester.edu/registrar/api/clusters/query"
    req = rq.post(url,
              data= { "academicDepartment": code,
                    "academicDiscipline": division})
    html = req.text
    soup =BeautifulSoup(html, "lxml")
    p =  json.loads(soup.find("p").text)
    Num_Cluster=len(p["results"])
    for cluster in range(0, Num_Cluster):
        # Cluster의 작은틀 숫자 
        Big_group = p["results"][cluster]["groups"]
        Table_Numbers_In_Cluster = len(Big_group)
        print("\n")
        print(p["results"][cluster]["header"]["clusterNumber"])

        for Num in range(0, Table_Numbers_In_Cluster):
            Number_Of_Courses=len(Big_group[Num]["courses"])
            if "groupDescription" not in p["results"][cluster]["groups"][Num]:
                break
            else:
                print(p["results"][cluster]["groups"][Num]["groupDescription"])
            for Num2 in range(0, Number_Of_Courses):

                print(Big_group[Num]["courses"][Num2]["courseNumber"])
```

```python
keys = list(Clusters_division.keys())
for key in keys:
    for i in Clusters_division[key]:
        Cluster(key, i)
```

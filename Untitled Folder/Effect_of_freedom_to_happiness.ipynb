```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("2019.csv")
```

## heatmap

#### I use heatmap to see correlation between features

```python
df = df.rename(columns = {"Freedom to make life choices":"Freedom"})
```

```python
cor = df.corr()
```

```python
sns.clustermap(cor, 
               annot = True,      
               cmap = 'RdYlBu_r'
              )
plt.savefig("heatmap.png")
```

#### Score does not have high relation with Freedom while score has high relation with Social support, GDP per capita, and healthy life expectancy

## line graph

```python
df = df.sort_values(by = "Freedom", ascending = False)
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x = "Freedom", y = "Overall rank")
plt.savefig("lineplot.png")
```

#### There is no country that has high freedom of choice than 0.7, so it is hard to check wether too high freedom leads to unhappy status

#### Although I could not answer my first question, I got some insight from this data. Line plot indicates  Freedom of choice is less than 0.2, people are unhappy. And it seems between 0.2 to 0.6 freedom is not important freature.
